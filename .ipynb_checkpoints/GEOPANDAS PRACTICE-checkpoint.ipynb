    
### Installation 
* Step one
 * Search for the anaconda prompt from your windows search bar.
 * Open and ensure that it reads base.
* Step two
 * Create a new virtual environment using the following command.
   * conda create -n your_environment_name
 * Activate this environment using the command:
     * conda activate your_environment_name
 * Add the following commands before installing geopandas:
     * conda config --env --add channels conda-forge
     * conda config --env --set channel_priority strict
 * Install the geopandas library as follows:
     * conda install python=3 geopandas
     * The reason why we install geopandas using this process is because geopandas using many dependences and this process ensures all of them are installed. 
* Step three
 * To ensure that the environment is present in the jupyter notebook kernels list, you need to install ipykernel. This is done as follows:
     * conda install -n your_environment_name ipykernel
     * python -m ipykernel install --user --name=our_environment_name
* Step four
 * Install jupyter notebook in your environment as follows:
     * conda install jupyter
 * You can then open the jupyter notebook from the anaconda prompt as follows
     * jupyter-notebook.
* Step five
 * On juoyter notebook, click on the New tab and on the downward menu, select your environment.

#### importing libraries

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Using the world GeoDataFrame for practice

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

#### Displaying the 1st 5 observations.

```python
world.head()
```

```python
world.plot();
```

#### Getting the name of the active geonmetry

```python
world.geometry.name
```

#### Renaming the geometry column to borders

```python
world = world.rename(columns = {'geometry':'borders'})
```

### Here, if you try to find the name of the active geometry column, you will get an error as shown below.
#### <font color = red> A GeoDataFrame keeps track of the active column by name, so if you rename the active geometry column, you must also reset the geometry:
#### This means that you have to now set the active geometry column as the column you have renamed to borders

```python
world.geometry.name
```

#### Setting borders as the active geometry column.

```python
world = world.set_geometry('borders')
```

```python
# Checking the active geometry column

world.geometry.name
```

#### Creating world centroids(The centroid is the centre point of the object)

```python
world['centroids'] = world.centroid
```

#### Making the centroids column to be the active geometry

```python
world = world.set_geometry('centroids')
```

#### Ploting the centroids

```python
world.plot();
```

#### Geopandas display options command

```python
gpd.options
```

#### Viewing the centroids: looking at the coordinates, the default number of decimals is 5. This is the default display precision for geographic coordinates.

```python
world['centroids'].head()
```

#### Changing the default number of decimals using the display_precision command, you can indicate the number of decimals you feel comfortable with. For example, lets change the display from 5 decimals to 9 decimals

```python
gpd.options.display_precision = 9
```

#### Now, viewing the 1st five entries of the cntroids column, you will notice that the number of decimal places changes from 5 to 9.

```python
world['centroids'].head()
```

### Adding a legend to the map
#### The way color maps are scaled can also be manipulated with the scheme option

```python
pip install mapclassify
```

```python
fig, ax = plt.subplots(1, 1)

Counties.plot(column='AREA', 
              ax=ax, 
              legend=True,
             legend_kwds={'label': "Area by County",
                        'orientation': "vertical"},
             cmap='Blues',
             scheme='quantiles');
```
